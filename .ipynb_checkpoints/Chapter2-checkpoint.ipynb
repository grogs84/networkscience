```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np

NS_PURPLE = "#8e44ad"
NS_GREEN  = "#2ecc71"

plt.rcParams.update({
    "axes.spines.top": False,
    "axes.spines.right": False,
    "axes.edgecolor": "0.3",
    "axes.labelsize": 14,
    "xtick.labelsize": 12,
    "ytick.labelsize": 12,
})

def draw_ns_graph_ax(G, ax, pos=None):
    if pos is None:
        pos = nx.spring_layout(G)
    nx.draw(
        G,
        pos=pos,
        node_color=NS_PURPLE,
        edge_color=NS_GREEN,
        node_size=800,
        width=4,
        with_labels=True,
        font_color="white",
        font_size=12,
        ax=ax,
    )
    ax.axis("off")

def plot_ns_degree_hist_ax(G, ax):
    degrees = [d for _, d in G.degree()]
    vals, counts = np.unique(degrees, return_counts=True)
    pk = counts / counts.sum()

    ax.bar(vals, pk, width=0.8, color=NS_PURPLE)
    ax.set_xlabel("k")
    ax.set_ylabel(r"$p_k$")
    ax.set_xlim(0, max(vals) + 1)
    ax.set_ylim(0, 1.05)
```

### 2.3 Degree, Average Degree and Degree Distribution
-----------------------------------------------------------

#### 2.3.1 Degree

Degree:
$L = \frac{1}{2} \sum_{i=1}^{N} k_i$

where $k_i$ is the degree of the $ith$ node in a network

```python
# nx.Graph.degree() give the node and degree for each node in the graph
L = 0
for node, node_degree in G.degree():
    L += node_degree

# take 1/2 because each edge connects to two nodes
L *= (1/2)
print(L)
```

```python
# networkx has a method to return the degree called number_of_edges
G.number_of_edges() == L
```

##### A graph can be represented as an adjacency matrix.

```python
print(nx.adjacency_matrix(G).todense())
```

```python
np.sum(nx.adjacency_matrix(G)) * .5
```

#### 2.3.2 Average Degree

Average Degree:
$<k> = \frac{1}{N} \sum_{i=1}^{N} k_i = \frac{2L}{N}$

where $k_i$ is the degree of the $ith$ node in a network

```python
# nx.Graph.degree() give the node and degree for each node in the graph
L = 0 # total edges
for node, node_degree in G.degree():
    L += node_degree
L *= (1./2)
N = G.number_of_nodes()
print(L*2/N)
```

```python
L = G.number_of_edges()
N = G.number_of_nodes()
print(f"Graph G has {L} edges, {N} nodes. The average degree is {2 * L/N}")
```

#### 2.3.3 Degree Distribution

##### The degree distribution is pk and is the probability that a randomly selected node in the network has degree k

$\sum_{i=1}^{inf} p_k = 1$

##### For a network with N nodes the degree distribution is the normalized histogram given by:

$p_k = \frac{N_k}{N}$

##### where ${N_k}$ is the number of degree-k nodes can be obtained from the degree distribution as ${N_k} = {Np_k}$

```python
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (2, 4), (3, 4)])
pos = {1: (2.5, 0.0), 2: (1.5, 0.0), 3: (1.0, -1.0), 4: (1.0, 1.0)}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

draw_ns_graph_ax(G, ax1, pos=pos)
plot_ns_degree_hist_ax(G, ax2)

fig.tight_layout()
plt.show()
```

```python
G = nx.watts_strogatz_graph(10, 2, 0)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

pos = nx.circular_layout(G)
draw_ns_graph_ax(G, ax1, pos=pos)
plot_ns_degree_hist_ax(G, ax2)

fig.tight_layout()
plt.show()
```
